```python

import pandas as pd
import string

df = pd.read_csv(r"F:\Compressed\archive_4\Ubuntu-dialogue-corpus\dialogueText_301.csv")
df.head()
df=df.dropna().reset_index(drop=True)
```

```python
df['clean_text']= df['text'].str.lower()
df.head()
```

```python
string.punctuation
```

```python

def remove_punctuations(text):
    punctuations= string.punctuation
    return text.translate(str.maketrans('','',punctuations))
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_punctuations(x))
df.head()
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
STOPWORDS= set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split()if word not in STOPWORDS])
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_stopwords(x))
df.head()
```

```python
from collections import Counter
word_count= Counter()
for text in df['clean_text']:
    for word in text.split():
        word_count[word]+=1

word_count.most_common(10)
```

```python
FREQUENT_WORDS = set(word for (word, wc) in word_count.most_common(10))
def remove_freq_words(text):
    return " ".join([word for word in text.split()if word not in FREQUENT_WORDS])
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_freq_words(x))
df.head()
```

```python
RARE_WORDS = set(word for (word, wc) in word_count.most_common()[:-1000:-1])
RARE_WORDS
```

```python
def remove_rare_words(text):
    return " ".join([word for word in text.split()if word not in RARE_WORDS])
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_rare_words(x))
df.head()
```

```python
import re
def remove_special_chars(text):
    text= re.sub('[^a-zA-Z0-9]', ' ',text)
    text= re.sub('\s+',' ',text)
    return text
```

```python
df['clean_text']= df['clean_text'].apply(lambda x: remove_special_chars(x))
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
df['stem_text']= df['clean_text'].apply(lambda x: stem_words(x))
df.head()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

def lemmatized_words(text):
    pos_text= pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word,wordnet_map.get(pos[0], wordnet.NOUN))for word , pos in pos_text])
```

```python
df['leammatized_text']= df['clean_text'].apply(lambda x: lemmatized_words(x))
df.head()
```

```python
def remove_url(text):
    return re.sub(r'https?://\s+|www\.\s+', '', text)
```

```python
df['leammatized_text']= df['leammatized_text'].apply(lambda x: remove_url(x))
df.head()
```

```python
def remove_html_tags(text):
    return re.sub(r'<.*?>','',text)
```

```python
df['leammatized_text']= df['leammatized_text'].apply(lambda x: remove_html_tags(x))
df.head()
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def correct_spellings(text):
    corrected_text= []
    misspelled_text = spell.unknown(text.split())
    for word in text.split():
        if word in misspelled_text:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
        
```

```python
 df['leammatized_text']= df['leammatized_text'].apply(lambda x: correct_spellings(x))
df.head()
```
